regressions with COVID-19 data

```
import numpy as np
import matplotlib.pyplot as plt  # To visualize
import pandas as pd  # To read data
from sklearn.linear_model import LinearRegression
```

```
url = 'https://raw.githubusercontent.com/nytimes/covid-19-data/master/us.csv'

covid = pd.read_csv(url, nrows = 106)
covid.head
```

```
datcl = covid;
datcl['date'] = pd.to_datetime(datcl['date'])
datcl.index = datcl['date']; 
datc = datcl.loc['2020-02-10':'2020-05-05']
#datc.plot(y = 'cases', use_index=True, title = 'covid-19 deaths per day', figsize = (10,5))
```

```
data = pd.read_csv("bigc.csv", header = 0) 
data['date'] = pd.to_datetime(data['date'])
data.index = data['date']; 
df = data;
save = df.groupby(df.index.date).count();
save.index = pd.to_datetime(save.index);
save.rename(columns={'id':'count'}, inplace=True)
cp = save[['count']].copy()
```

```
l = cp.loc['2020-02-10':'2020-05-05']
fin =np.log(l['count'])
cdat = np.log(datc['cases'].copy())
cat = pd.concat([cdat, fin], axis=1)
X = cat.iloc[:, 0].values.reshape(-1, 1)  # values converts it into a numpy array
Y = cat.iloc[:, 1].values.reshape(-1, 1)  # -1 means that calculate the dimension of rows, but have 1 column
linear_regressor2 = LinearRegression()  # create object for the class
linear_regressor2.fit(X, Y)  # perform linear regression
Y_pred2 = linear_regressor2.predict(X)  # make predictions
```

```
from matplotlib.pyplot import figure
plt.scatter(X, Y)
plt.plot(X, Y_pred2, color='red')
plt.xlabel("Covid-19 cases in US")
plt.ylabel("#buildthewall tweets")
plt.title("Linear Regression of log cases vs #buildthewall tweets")
plt.figure(figsize = (16, 7))
plt.show()
```

```
cat.plot(x = 'cases', y = 'count', title = '#buildthewall tweets per US COVID-19 case',figsize = (10,5))
```
